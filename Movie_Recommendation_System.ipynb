```python
import requests_with_caching
import json


############### Using Tastedive API Call to fetch movie data. ########################

def get_movies_from_tastedive(str):
    parameter={'q': str,'type': 'movies','limit':5}
    baseurl="https://tastedive.com/api/similar"
    res=requests_with_caching.get(baseurl,params=parameter)
    print(res.url)
    
    return(res.json()) 

###################### Extracting Movie Titles from the Data returned by Tastedive API. ############ 

def extract_movie_titles(get_movies_from_tastedive):
    inpt=get_movies_from_tastedive
    new_list=[]
    it2=inpt['Similar']['Results']
    
    for ite in it2:
        new_list.append(ite['Name'])
    return new_list

###### It takes a list of movie titles as input. It gets five related movies for each from TasteDive,####
###### extracts the titles for all of them, and combines them all into a single list. ###################                    


def get_related_titles(new_list):
    good_list=[]
    x=[]
    for item in new_list:
        x=extract_movie_titles(get_movies_from_tastedive(item))
        for itm in x:
            if itm not in good_list:
                good_list.append(itm)
    return good_list
 
##### Getting Movie Data such as its rating on Rotten Tomatoes, Review etc. ########
       
def get_movie_data(str):
    parameter={'t': str,'r': 'json'}
    baseurl="http://www.omdbapi.com/"
    res=requests_with_caching.get(baseurl,params=parameter)
    #print(res.text)
    #print(res.url)
    return res.json()

##### Extracts the Rating for each movie in the Data Set returned by OMDB API #######

def get_movie_rating(dic):
    ranking = dic['Ratings']
    for dic_item in ranking:
        if dic_item['Source'] == 'Rotten Tomatoes':
            return int(dic_item['Value'][:-1])
    return 0

###### It takes a list of movie titles as an input. ########
###### It returns a sorted list of related movie titles as output,#######
####### up to five related movies for each input movie title.###########


def get_sorted_recommendations(input_list):
    unsorted_list=get_related_titles(input_list)
    movie_scores=[]
    final_dict={}
    for ite in unsorted_list: 
        if ite not in final_dict:
            
            final_dict[ite]=get_movie_rating(get_movie_data(ite))
    sorted_dictionary=sorted(final_dict,key=lambda x :(final_dict[x],final_dict.keys()),reverse=True)
        
                                             
    return sorted_dictionary                                     
                                             
get_sorted_recommendations(["Bridesmaids", "Sherlock Holmes"])

```
